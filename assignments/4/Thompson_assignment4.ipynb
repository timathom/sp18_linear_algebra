1.

The coefficient of the linear term m is equal to rise over run, which is the change in the y coordinate divided by the change in the x coordinate.

y_1 - y_0 = m(x_1 - x_0)

Solving for m, we have:
m = y_1 - y_0 / x_1 - x_0

If two points are the same, there will be no slope, as in the horizontal line of the x-axis, or the vertical line of
the y-axis.

2.

x^2+y^2=1 is the formula for a circle with radius 1.
The first part of the Python code, `concatenate((x, x))`, corresponds to x^2 (two arrays concatenated). The second part of the code, `concatenate((sqrt(1-x*x), -sqrt(1-x*x)))`, expresses y^2 in terms of x. Solving the equation for y, we get the square root of 1-x^2. The square root can have either a positive or negative value, so we need two arrays for y in terms of x: `sqrt(1-x*x)` and `-sqrt(1-x*x)`.

However, when it is passed another point `(x1, y1)`, it creates a secant line that approximates the tangent as the limit of h approaches 0.

4.

The derivative of the unit circle at (1/sqrt(2), 1/sqrt(2)) is -1.

```python
data = np.linspace(-2, 2, 1001)
x2 = np.sqrt(2)/2


def point(x, y): plt.plot(x, y, 'o', label="$(" + str(x) + "," + str(y) + ")$")


def points_form(x0, y0, x1, y1, x, start, stop, dim):
    point_slope_form(x0, y0, (y1-y0)/(x1-x0), x, start, stop, dim)


def point_slope_form(x0, y0, m, x, start, stop, dim):
    plt.plot(x,
             np.linspace(m * (start - x0) + y0, m * (stop - x0) + y0, dim),
             label="$y=" + str(m) + "x+" + str(m * (-x0) + y0) + "$")


plt.plot(data, np.zeros(1001), label="$x$-axis")
plt.plot(np.zeros(1001), data, label="$y$-axis")
plt.plot(np.concatenate((data, data)), np.concatenate((np.sqrt(1-data**2), -np.sqrt(1-data**2))), label="$x^2+y^2=1$")
point_slope_form(x2, x2, -1., data, -2, 2, 1001)
plt.ylim(-2, 2)
plt.xlim(-2, 2)
point(x2, x2)

plt.legend()
plt.show()
```

The tables in the lecture notes are inconsistent, poorly worded, and difficult to compare. In general, the parabola graphed by x^2 shares some properties of both the unit circle table and the table for f(x)=2x^3-3x^2+9.

|               | x-axis                             | \(y = 2.25x + -1.25\)                      | \(y = 0.0x + 3.0625\)                                       |
| ------------- | ---------------------------------- | ------------------------------------------ | ----------------------------------------------------------- |
| Approximation | Closely approximates the parabola. | Does not closely approximate the parabola. | Begins to closely approximate the parabola.                 |
| Intersection  | Touches the parabola once.         | Touches the parabola twice.                | (1.0, 1.0) and (1.25, 1.5625)                               |
| Sides         | The parabola is on one side of it. | The parabola is on both sides of it.       | The parabola is on both sides of it.                        |
| Type          | Tangent line                       | Secant line                                | Secant line                                                 |

```python
def g(x): return x**2


plt.plot(data, g(data))

point_slope_form(0., 0., 0., data, -2, 2, 1001)
points_form(1., g(1.), 1.25, g(1.25), data, -2, 2, 1001)
points_form(-1.75, g(-1.75), 1.75, g(1.75), data, -2, 2, 1001)

plt.plot(data, np.zeros(1001), label="$x$-axis")
plt.plot(np.zeros(1001), np.linspace(-4, 4, 1001), label="$y$-axis")

point(0., 0.)
point(1., 1.)
point(-1.75, g(-1.75))
point(1.75, g(1.75))
point(1.25, g(1.25))

plt.xlim(-2, 2)
plt.ylim(-4, 4)
plt.legend()

plt.show()
```

This is not a tangent line. The slope is equal to 1 or very close to one.


|               | \(y = 1.0x + 1.0\)                      | \(y = 0.9999999999999998x + 1.0\)       | \(y = 1.0000000000000009x + 1.0\)       |
| ------------- | --------------------------------------- | --------------------------------------- | --------------------------------------- |
| Approximation | Does not closely approximate the curve. | Does not closely approximate the curve. | Does not closely approximate the curve. |
| Intersection  | Touches the curve once.                 | Touches the curve once.                 | Touches the curve once.                 |
| Sides         | The curve is on one side of it.         | The curve is on both sides of it.       | The curve is on both sides of it.       |
| Type          | Secant line                             | Secant line                             | Secant line                             |

```python
def e_x2(x):
    return np.e**(-x**2)


plt.plot(data, e_x2(data))
point(0., 1.)
for h in [0.5, 0.2, 0.1]:
    points_form(0., 1, 0. + h, 1. + h, data, -2, 2, 1001)

    point(0. + h, 1. + h)

# plt.legend()
plt.show()
```
