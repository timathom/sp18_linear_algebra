```python
data = np.linspace(-2, 2, 101)
data2 = np.linspace(-1, 5, 101)
```

1.

The function f(x) = 1 - |x| has a maximum at x = 0, but the derivative of the function is not equal to 0 at that point. It is undefined. (Taken from https://math.stackexchange.com/questions/605158/can-a-derivative-be-undefined-at-a-local-maxima)

```python
def f(x):
    return 1 - np.abs(x)


plt.plot(data, f(data))
plt.show()
```

2.

The derivative of the function f(x) = x^3 is 0 at f'(0), but the point at (0, 0) is flat: it is neither a maximum nor a minimum. (Taken from https://web.archive.org/web/20190424060644/https://www.whitman.edu/mathematics/calculus_online/section05.01.html)

```python
def f2(x): return x**3


print("Derivative at x = 0: ", Derivative(f2)(0.0))

plt.plot(data, f2(data))
plt.show()
```

3.

The function x^3 - 6x^2 + 9x + 2 on the interval (-1, 5) has a local maximum at (1, 6), but it does not have a global maximum. (Taken from https://web.archive.org/web/20180724175621/http://www2.gcc.edu/dept/math/faculty/BancroftED/buscalc/chapter2/section2-7.php)

```python
def f3(x): return x**3 - (6 * x**2) + (9 * x) + 2


plt.plot(data2, f3(data2))
plt.show()
```

4.

```python
def f4(x): return (6 * x**2) - (6 * x)


y = (6 * x**2) - (6 * x)

print("Derivative: ", y.diff(x))

cp = solve(y.diff(x), x)

print("Critical point: ", cp)

for p in cp:
    print(y.diff(x, 2).subs(x, p))

plt.plot(data2, f4(data2))
plt.show()
```
