The formula for compound interest is (1 + 1/n)^2

```python
def compound_interest(v, n):
    return [v * (1 + 1.0/i)**i for i in range(1, n + 1, 2)]


e = compound_interest(1, 10000)
print(e[len(e) - 1])

# Plot code adapted from: https://gist.github.com/Quackmatic/0e6073ecce62df2db467
# Rate of compound interest converges to e (2.718...)

plt.plot(e)
plt.plot([0, 6], [2.7183, 2.7183], "g--")  # Show the actual value of e with a dotted green line
plt.xlim([0, 6])
plt.ylim([0, 6])
plt.show()
```

8.

Plot the graph for $e^{-x^2}$

```python
x5 = np.linspace(-2, 2, 101)


def f_q8(x):
    return np.e**(-x**2)


plt.plot(x5, f_q8(x5), label="$e^{-x^2}$")
plt.legend()
plt.show()
```

Function `f_q8(x)` ($e^{-x^2}$) does not have a root because it is asymptotic--it never intersects with zero.

The function is increasing when x < 1, and it is decreasing with x > 1.

If the domain of the function f_q8(x) is (-inf, inf), then it has no inverse function. To have an inverse function, the domain of f_q8(x) must be restricted to [0, inf), and the range of its inverse must also be restricted to [0, inf).

```python
def f_inv_q8(x):
    return np.sqrt(-(np.log(x)))


inverse = f_inv_q8(x5)

plt.plot(x5, f_q8(x5), label="$e^{-x^2}$")
plt.plot(x5, inverse, label="$\sqrt{-\ln(x)}$")
plt.plot(x5, x5, label="$y=x$")
plt.ylim((0, 2))
plt.xlim((0, 2))
plt.legend()
plt.show()
```

Exponential functions with base 0 define a simple constant function, and expressions with a negative base may not define a real number [1]. The graph for base 0 shows a constant function from (1, 0) and (0, 2). The graph for -1 cannot be generated because it is invalid.

[1] http://wmueller.com/precalculus/families/1_21.html

```python
x6 = np.linspace(0, 2, 101)


def f_zero(x):
    b = 0
    return b**x


def f_neg_one(x):
    b = -1
    return b**x


plt.plot(x6, f_zero(x6))
plt.show()

plt.plot(x6, f_neg_one(x6))
plt.show()
```
